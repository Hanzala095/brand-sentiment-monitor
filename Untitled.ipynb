```python
# Step 1: Install Required Libraries
!pip install -q transformers pandas matplotlib seaborn

# Step 2: Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from transformers import pipeline

# Step 3: Load Dataset
df = pd.read_csv('Tweets.csv')


# Step 4: Preprocessing
df = df[["airline", "text", "airline_sentiment", "tweet_created"]]
df = df.rename(columns={
    "airline": "Brand",
    "text": "Tweet",
    "airline_sentiment": "Sentiment",
    "tweet_created": "Date"
})

# Step 5: Ensure datetime format
df["Date"] = pd.to_datetime(df["Date"])
df["DateOnly"] = df["Date"].dt.date

# Step 6: Sentiment Distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="Sentiment", palette="Set2")
plt.title("Overall Sentiment Distribution")
plt.grid(True)
plt.show()

# Step 7: Create Sentiment Trend Table
trend = df.groupby(["DateOnly", "Brand", "Sentiment"]).size()
trend = trend.unstack(level="Sentiment").fillna(0)  # Columns: Sentiment, Index: Date + Brand

# Step 8: Plot Trends for Each Brand
brands = df["Brand"].unique()

for brand in brands:
    if brand not in trend.index.get_level_values(1):
        continue  # skip if brand not in trend index

    sentiment_data = trend.xs(brand, level=1)
    sentiment_data.plot(kind='line', title=f"Sentiment Trend for {brand}", figsize=(10, 5))
    plt.xlabel("Date")
    plt.ylabel("Tweet Count")
    plt.legend(title="Sentiment")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Step 9: Anomaly Detection Function
def detect_anomalies(trend_data, threshold_percent=50):
    print("----- Anomaly Alerts -----")
    for brand in trend_data.index.get_level_values(1).unique():
        brand_data = trend_data.xs(brand, level=1)
        if len(brand_data) < 2:
            print(f" Not enough data points to detect anomalies for {brand}.")
            continue
        for sentiment in brand_data.columns:
            prev_value = None
            for date, value in brand_data[sentiment].items():
                if prev_value is not None and prev_value > 0:
                    percent_change = ((value - prev_value) / prev_value) * 100
                    if abs(percent_change) >= threshold_percent:
                        direction = "Increase" if percent_change > 0 else "Decrease"
                        print(f" {direction} in '{sentiment}' sentiment for {brand} on {date}: {percent_change:.1f}%")
                prev_value = value

# Step 10: Run Anomaly Detection
detect_anomalies(trend)

# Step 11: Optional — Named Entity Recognition on Sample Tweet
ner_pipe = pipeline("ner", model="dslim/bert-base-NER", aggregation_strategy="simple")
sample_tweet = df["Tweet"].iloc[0]
ner_results = ner_pipe(sample_tweet)

print("\nNamed Entities in Sample Tweet:")
for entity in ner_results:
    print(f"{entity['word']} ({entity['entity_group']}) - Score: {round(entity['score'], 3)}")
```

```python
# Step 1: Install Required Libraries
!pip install -q tweepy transformers pandas matplotlib seaborn

# Step 2: Imports
import os
import tweepy
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from transformers import pipeline

# Step 3: Setup Twitter API
BEARER_TOKEN = "YOUR_TWITTER_BEARER_TOKEN"
client = tweepy.Client(bearer_token=BEARER_TOKEN)

# Step 4: Setup NLP Models
os.environ["TRANSFORMERS_NO_TF"] = "1"
sentiment_pipe = pipeline("text-classification", model="finiteautomata/bertweet-base-sentiment-analysis")
ner_pipe = pipeline("ner", model="dslim/bert-base-NER", aggregation_strategy="simple")

# Step 5: Collect Tweets Function
def collect_brand_data(brand_query, brand_name, max_results=30):
    tweets_data = []

    print(f"\n Collecting tweets for: {brand_name}")
    try:
        response = client.search_recent_tweets(
            query=brand_query,
            max_results=max_results,
            tweet_fields=["created_at", "author_id", "text", "public_metrics", "lang"]
        )
    except tweepy.TooManyRequests:
        print("Rate limit exceeded. Try again later.")
        return pd.DataFrame()

    if not response.data:
        print(f"No tweets found for {brand_name}")
        return pd.DataFrame()

    for tweet in response.data:
        if tweet.lang == "en":
            text = tweet.text
            sentiment_result = sentiment_pipe(text)[0]
            ner_result = ner_pipe(text)
            entities = ", ".join(set([e['word'].replace("##", "") for e in ner_result]))

            tweets_data.append([
                tweet.created_at,
                tweet.author_id,
                brand_name,
                text,
                tweet.public_metrics['like_count'],
                tweet.public_metrics['retweet_count'],
                sentiment_result['label'],
                round(sentiment_result['score'], 3),
                entities
            ])

    return pd.DataFrame(tweets_data, columns=[
        "Date", "User ID", "Brand", "Tweet", "Likes", "Retweets",
        "Sentiment", "Sentiment Score", "Entities"
    ])

# Step 6: Collect Data from Twitter
brands = {
    "Nestle": "Nestle OR @Nestle OR #Nestle lang:en -is:retweet",
    "Pepsi": "Pepsi OR @Pepsi OR #Pepsi lang:en -is:retweet"
}

dfs = []

for brand, query in brands.items():
    df = collect_brand_data(query, brand, max_results=30)
    dfs.append(df)

all_data = pd.concat(dfs, ignore_index=True)

# Step 7: Save to CSV
if not all_data.empty:
    all_data.to_csv("live_brand_sentiment.csv", index=False, encoding="utf-8-sig")
    print(" Tweets saved to live_brand_sentiment.csv")
else:
    print(" No tweets collected.")
```
